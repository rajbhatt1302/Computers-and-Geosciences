```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.lines as mlines
from sklearn.metrics import r2_score
import scipy.stats
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
df = pd.read_csv('master_subset.csv')
display(df)
x=df.lst
y=df.predicted2p
z=df.predicted3p
```

```python
r2 = r2_score(x,z)
print('r2 score for perfect model is', r2)
r = scipy.stats.pearsonr(x, z)  
print('r score for perfect model is', r)
mape=mean_absolute_percentage_error(x,z)
print(mape,'mape')
print(np.sqrt(mape),'rmspe')

```

```python
sns.set_theme(color_codes=True)
fig = plt.figure(figsize=(25,15))

sns.regplot(x,y,color='red',marker="o",scatter_kws=dict(s=140, color='',edgecolor="black", linewidth=1.5, alpha=0.95),fit_reg=True,truncate=True);
plt.xlabel('Satellite LST (°C)',fontsize=25,labelpad=17)
plt.ylabel('Predicted LST (°C)',fontsize=25,labelpad=17)
plt.xticks((np.arange(29,35,1)),fontsize=28)
plt.yticks(np.arange(29,33,1),fontsize=28)

plt.text(30,32,'$R^2$ : 0.91',fontsize=28)
plt.text(30,31.5,'r : 0.98',fontsize=28)
plt.text(30,31,'RMSPE : 0.08',fontsize=28)
plt.text(30,30.5,'MAPE : 0.006',fontsize=28)



plt.plot()

```
